#AutoML

AutoML are libraries that automate the process of algorithm selection, hyperparameter tuning, and feature engineering.

```python
from tpot import TPOTClassifier
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
from sklearn.model_selection import train_test_split
```

```python
cancer=load_breast_cancer()
```

```python
x_train, X_test, y_train , y_test = train_test_split(cancer.data,cancer.target,test_size =0.2 , random_state=42)
```

```python
topt= TPOTClassifier(generations=5, population_size=20 ,verbosity=2,random_state=42)
```

```python
topt.fit(x_train, y_train)
```

```python
accuracy=topt.score(X_test, y_test)
```

```python
print(accuracy)
```

```python
topt.export('model_best.py')
```
