## Imputing Missing Values

```python
import numpy as np
num=data.select_dtypes(include=np.number).columns
data[num]=data[num].apply(pd.to_numeric,errors='coerce')
data[num]=data[num].fillna(data[num].mean())
data[num]
```

```python
cat_columns=data.select_dtypes(include='object').columns
data[cat_columns]=data[cat_columns].fillna(data[cat_columns].mode().iloc[0])
```

```python
model=setup(data,target='Genre')
compare_models()
```
